```python
import pandas as pd
```

```python
# alterando a Data que est√° como Objeto ou seja Strg e transformando em DateTime
results= pd.read_csv('/content/results.csv', parse_dates=['date'])
```

```python
results
```

```python
#Corfirmando que agora √© DataTime

results.info()
```

```python
#Filtrar por ano
results['date'].dt.year
```

```python
# Extrair somente dados da Copa criando uma nova vari√°vel
results_fifa=results[results['tournament']== 'FIFA World Cup']
```

```python
results_fifa
```

```python
#Vamos criar uma coluna Ano atribuindo a uma coluna do dataframe
results_fifa['year']= results_fifa['date'].dt.year
```

```python
home_score=results_fifa[['home_team', 'home_score', 'year']]
away_score=results_fifa[['away_team', 'away_score', 'year']]
```

```python
home_score
```

```python
home_score=results_fifa[['home_team', 'home_score', 'year']].rename({'home_team': 'team','home_score': 'score'}, axis='columns')
away_score=results_fifa[['away_team', 'away_score', 'year']].rename({'away_team': 'team','away_score': 'score'}, axis='columns')
```

```python
home_score
```

```python

score_fifa= pd.concat([home_score, away_score]).reset_index(drop=True).fillna(0)
```

```python
score_fifa
```

```python
score_fifa.groupby(['team', 'year']).score.sum().unstack().fillna(0)
```

```python
score_fifa.groupby(['team', 'year']).score.sum().unstack().fillna(0).cumsum(axis=1)
```

```python
score_fifa_year=score_fifa.groupby(['team', 'year']).score.sum().unstack().fillna(0).cumsum(axis=1).reset_index()
```
